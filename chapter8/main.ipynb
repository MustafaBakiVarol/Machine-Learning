# UNSUPERVISED LEARNING

Bugün Makine Öğrenimi uygulamalarının çoğu denetimli öğrenmeye dayansa da (ve bu nedenle yatırımların çoğu buraya yöneliyor), mevcut verilerin büyük çoğunluğu etiketsizdir: girdi özellikleri X'e sahibiz, ancak etiketler y elimizde yok. Bilgisayar bilimcisi Yann LeCun ünlü bir şekilde şunları söylemiştir: “Eğer zekâ bir pasta olsaydı, denetimsiz öğrenme pastanın kendisi olurdu, denetimli öğrenme pastanın kreması olurdu ve pekiştirmeli öğrenme pastanın üzerindeki kiraz olurdu.” Başka bir deyişle, denetimsiz öğrenmede henüz tam olarak kullanmaya başlamadığımız büyük bir potansiyel vardır. Diyelim ki üretim hattındaki her ürünün birkaç fotoğrafını çekip hangi ürünlerin hatalı olduğunu tespit edecek bir sistem oluşturmak istiyorsunuz. Fotoğrafları otomatik olarak çekebilecek bir sistem oluşturmak oldukça kolay olabilir ve bu her gün size binlerce fotoğraf verebilir. Ardından sadece birkaç hafta içinde oldukça büyük bir veri seti oluşturabilirsiniz. Ama bekleyin, etiket yok! Eğer bir öğenin hatalı olup olmadığını tahmin edecek normal bir ikili sınıflandırıcı eğitmek istiyorsanız, her bir fotoğrafı “hatalı” veya “normal” olarak etiketlemeniz gerekecek. Bu genellikle insan uzmanların oturup tüm fotoğrafları manuel olarak incelemesini gerektirir. Bu uzun, maliyetli ve sıkıcı bir görevdir, bu yüzden genellikle yalnızca mevcut fotoğrafların küçük bir alt kümesi üzerinde yapılır. Sonuç olarak, etiketli veri seti oldukça küçük olacak ve sınıflandırıcının performansı hayal kırıklığı yaratacaktır. Dahası, şirket ürünlerinde herhangi bir değişiklik yaptığında, süreç baştan başlamak zorunda kalacaktır. Algoritmanın her fotoğrafı etiketlemeden etiketsiz veriyi kullanabilmesi harika olmaz mıydı? İşte denetimsiz öğrenme burada devreye giriyor.

Clustering 
    - Amaç, benzer örnekleri kümeler halinde gruplamaktır. Kümelenme, veri analizi, müşteri segmentasyonu, öneri sistemleri, arama motorları, görüntü segmentasyonu, yarı denetimli öğrenme, boyut indirgeme ve daha fazlası için harika bir araçtır.

Anomaly detection
    -  Amacı, "normal" verinin nasıl göründüğünü öğrenmek ve ardından bunu üretim hattındaki hatalı ürünler veya bir zaman serisindeki yeni bir trend gibi anormal durumları tespit etmek için kullanmaktır.

 Density estimation
    - Bu, veri setini üreten rastgele sürecin olasılık yoğunluk fonksiyonunu (PDF) tahmin etme görevidir. Yoğunluk tahmini, genellikle anomali tespiti için kullanılır: çok düşük yoğunluklu bölgelerde bulunan örnekler muhtemelen anomalilerdir. Ayrıca veri analizi ve görselleştirme için de faydalıdır.
