Anomaly detection
    -  Amacı, "normal" verinin nasıl göründüğünü öğrenmek ve ardından bunu üretim hattındaki hatalı ürünler veya bir zaman serisindeki yeni bir trend gibi anormal durumları tespit etmek için kullanmaktır.

 Density estimation
    - Bu, veri setini üreten rastgele sürecin olasılık yoğunluk fonksiyonunu (PDF) tahmin etme görevidir. ## Clustering 

- Müşteri segmentasyonu için, müşterilerinizi satın almalarına ve web sitenizdeki etkinliklerine göre gruplandırabilirsiniz. Bu, müşterilerinizin kim olduğunu ve neye ihtiyaç duyduklarını anlamak için faydalıdır, böylece ürünlerinizi ve pazarlama kampanyalarınızı her segmente uyarlayabilirsiniz. Örneğin, müşteri segmentasyonu, aynı kümeye ait diğer kullanıcıların beğendiği içerikleri önermek için öneri sistemlerinde faydalı olabilir.

- Veri analizi için, yeni bir veri setini analiz ederken, bir kümeleme algoritması çalıştırmak ve ardından her kümeyi ayrı ayrı analiz etmek faydalı olabilir.

- Boyut indirgeme tekniği olarak, bir veri kümesi kümeleme işlemine tabi tutulduğunda, genellikle her bir örneğin her bir küme ile olan uyumunu ölçmek mümkündür (uyum, bir örneğin bir kümeye ne kadar iyi uyduğunu gösteren herhangi bir ölçüttür). Daha sonra her bir örneğin özellik vektörü x, küme uyumları vektörü ile değiştirilebilir. Eğer k tane küme varsa, bu vektör k boyutludur. Bu vektör genellikle orijinal özellik vektöründen çok daha düşük boyutludur, ancak daha ileri işlemler için yeterli bilgiyi koruyabilir.

- Anomali tespiti (aykırı tespit olarak da adlandırılır) için, tüm kümelere düşük bir yakınlığa sahip herhangi bir örnek muhtemelen bir anomali olarak değerlendirilir. Örneğin, web sitenizin kullanıcılarını davranışlarına göre kümelendirdiyseniz, saniye başına alışılmadık sayıda istek gibi olağandışı davranış sergileyen kullanıcıları tespit edebilirsiniz. Anomali tespiti, özellikle üretimdeki kusurları veya dolandırıcılıkları tespit etmekte faydalıdır.

- Yarı denetimli öğrenme için sadece birkaç etiketiniz varsa, kümeleme yapabilir ve etiketleri aynı kümedeki tüm örneklere yayabilirsiniz. Bu teknik, sonraki denetimli öğrenme algoritması için mevcut etiket sayısını büyük ölçüde artırabilir ve böylece performansını iyileştirebilir. 

- Arama motorları Bazı arama motorları, referans bir görsele benzer görselleri aramanıza olanak tanır. Böyle bir sistem kurmak için, öncelikle veritabanınızdaki tüm görsellere bir kümeleme algoritması uygulamanız gerekir; benzer görseller aynı kümede toplanır. Daha sonra bir kullanıcı bir referans görsel sağladığında, yapmanız gereken tek şey eğitilmiş kümeleme modelini kullanarak bu görselin kümesini bulmak ve ardından bu kümedeki tüm görselleri döndürmektir.
