Bir Karar Ağacını Eğitmek ve Görselleştirmek
Karar Ağaçlarını anlamak için, bir tane oluşturalım ve tahminleri nasıl yaptığını inceleyelim. Aşağıdaki kod, iris veri seti üzerinde bir DecisionTreeClassifier modeli eğitir:

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris() 
X = iris.data[:, 2: ]
y = iris.target 

tree_clf = DecisionTreeClassifier(max_depth = 2)
tree_clf.fit(X, y)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(tree_clf, out_file = "iris_tree.png", feature_names = iris.feature_names[2:], class_names = iris.target_names, rounded = True, filled = True )
```

```python
tree_clf.predict_proba([[5, 1.5]])
```

```python
tree_clf.predict([[5, 1.5]])
```

Algoritma, ilk olarak eğitim setini tek bir özellik k ve bir eşik tk kullanarak iki alt kümeye ayırarak çalışır (örneğin, “taç yaprağı uzunluğu ≤ 2,45 cm”). k ve tk nasıl seçilir? Algoritma, en saf alt kümeleri üreten (boyutlarına göre ağırlıklandırılmış) (k, tk) çiftini arar. Denklem 6-2, algoritmanın minimize etmeye çalıştığı maliyet fonksiyonunu verir.

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeClassifier(max_depth = 2)
tree_reg.fit(X, y)
```
